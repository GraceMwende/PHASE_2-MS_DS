### 1. Find the employees who's salary is more than the average salary earned by all employees
####    step1: Find avg salary
####    step2: Filter employees based on the above result

```python
#Avg salary to be used as subquery
q = '''
SELECT AVG(salary) 
from employee
'''
pd.read_sql(q,conn)
```

```python
q = '''
SELECT *  --outer query/main query
    FROM employee where salary > (
    SELECT AVG(salary) 
    from employee --subquery/inner query
)
'''
pd.read_sql(q,conn)
```

### b)Multiple row subquery
- Subquery which returns multiple column and multiple rows
- return only 1 column with multiple rows

### c)Correlated subquery
Subquery which is related to the outer query- not good for db with multiple records inner join are preferred but this coorelated sub can help solve complex statements
### i)Find the employees in each department who earn more than the average salary in that department

```python
#select avg(salary) from employee where dept_name = "specific_debt"
```

```python
q = '''
select * 
from employee e1 
where salary > 
    (
    select avg(salary) 
    from employee e2
    where e2.dept_name =  e1.dept_name --subquery dependent on the outer and cant be executed alone
    )
'''
pd.read_sql(q,conn)
```

### ii)Find Department who do not have any employees

```python
q = '''
SELECT * 
FROM department d
WHERE not exists (
select 1 from employee e where e.dept_name = d.dept_name
)
'''
pd.read_sql(q,conn)
```

### NESTED SUBQUERIES
### Q: Find stores who's sales were better than the average sales across all the stores
1) Find the total sales for each store
2) Find avg sales for all the stores
3) Compare 1 & 2

```python
q ='''
SELECT * FROM sales
'''
pd.read_sql(q,conn)
```

#### step 1:Total sales for each store

```python
q = '''select store_name,sum(price) as total_sales
from sales
group by store_name 
'''
pd.read_sql(q,conn)
```

### step2: Avg sales for all the stores

```python
q = '''select avg(total_sales) from 
(
select store_name,sum(price) as total_sales
from sales
group by store_name 
)
'''
pd.read_sql(q,conn)
```

### step 3: Compare 1 & 2

```python
q = '''
SELECT * from (
    select store_name,sum(price) as total_sales
    from sales
    group by store_name) sales
JOIN  (select avg(total_sales) as sales from 
(
    select store_name,sum(price) as total_sales
    from sales
    group by store_name) x)avg_sales
    on sales.total_sales > avg_sales.sales
'''
pd.read_sql(q,conn)
```

### Better query using with clause

```python
q = '''
with sales as
	(select store_name, sum(price) as total_sales
	 from sales
	 group by store_name)
select *
from sales
join (select avg(total_sales) as avg_sales from sales) avg_sales
	on sales.total_sales > avg_sales.avg_sales;

'''
pd.read_sql(q,conn)
```
